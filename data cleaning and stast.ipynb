### **Q1 --> seeing that the men and women avg cholestrol are they coming from the sample population**
- signifance level = 0.05 


**Hypothesis Testing**
- H0 - Null hypothesis --> (assuming) both are coming from the same population 
- H1 - alternate hypothesis  --> claming both are not coming from the same population 


```python
from statsmodels.stats.weightstats import ztest

men = df[df['sex'] == 1 ]['cholesterol']
women  = df[df['sex'] == 0 ]['cholesterol']
alpha = 0.05


z_stastistics, p_value  = ztest(men, women)

print(f'z_stastistics = {z_stastistics}' )
print(f'P_value = {p_value}' )


if p_value < alpha:
    print("P-value < significance level")
    print("We reject the null hypothesis.")
    print("There is strong evidence that men and women do not come from the same population (cholesterol levels differ).")
else:
    print("P-value >= significance level")
    print("We fail to reject the null hypothesis.")
    print("There is not enough evidence to say men and women have different cholesterol levels.")
```

### **Q2 --> seeing that the heart_disease and not heart diseases  avg cholestrol are they coming from the sample population**
- signifance level = 0.05 


**Hypothesis Testing**
- H0 - Null hypothesis --> (assuming) both are coming from the same population 
- H1 - alternate hypothesis  -->claming both are not  coming from the same population 


```python
from statsmodels.stats.weightstats import ztest

chol_with_disease = df[df['heart_disease'] == 1 ]['cholesterol']
chol_without_disease  = df[df['heart_disease'] == 0 ]['cholesterol']
alpha = 0.05


z_stastistics, p_value  = ztest(chol_with_disease, chol_without_disease)

print(f'z_stastistics = {z_stastistics}' )
print(f'P_value = {p_value}' )


if p_value < alpha:
    print("P-value < significance level")
    print("We reject the null hypothesis.")
    print("There is strong evidence that chol_with_disease and chol_without_disease do not come from the same population (cholesterol levels differ).")
else:
    print("P-value >= significance level")
    print("We fail to reject the null hypothesis.")
    print("There is not enough evidence to say chol_with_disease and chol_without_disease have different cholesterol levels. they are coming from the same poulation ")
```

### **Q3--> is there any signinficance difference between the cholesterol level between the different chest pain?** 


```python
df['chest_pain'].value_counts()
```

| Value | Meaning              | Description                                                               |
| ----- | -------------------- | ------------------------------------------------------------------------- |
| 0     | **Typical angina**   | Chest pain related to physical exertion or stress, relieved by rest/meds. |
| 1     | **Atypical angina**  | Chest pain not clearly related to exertion or stress.                     |
| 2     | **Non-anginal pain** | Chest pain not related to the heart (e.g., gastrointestinal or muscular). |
| 3     | **Asymptomatic**     | No chest pain, but other symptoms (e.g., shortness of breath, fatigue).   |


```python
from scipy.stats import f_oneway
```

```python
cholesterol_0 = df[df['chest_pain'] == 0]['cholesterol']
cholesterol_1 = df[df['chest_pain'] == 1]['cholesterol']
cholesterol_2 = df[df['chest_pain'] == 2]['cholesterol']
cholesterol_3 = df[df['chest_pain'] == 3]['cholesterol']


f_stats , p_value = f_oneway(cholesterol_0,cholesterol_1,cholesterol_2,cholesterol_3)


print(f"F-statistic: {f_stats}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: At least one group has a different mean cholesterol level.")
else:
    print("Fail to reject the null hypothesis: No significant difference in mean cholesterol between chest pain types.")
```
